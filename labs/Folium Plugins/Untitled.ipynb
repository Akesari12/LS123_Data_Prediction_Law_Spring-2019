```python
example_map1 = folium.Map([39.83, -98.59], zoom_start=6)

data = (np.random.normal(size=(100, 2)) *
        np.array([[1, 1]]) +
        np.array([[39.83, -98.59]])).tolist()

for row in data:
    folium.Circle(
            radius = 5,
        location = (row[0], row[1]),
        color = 'green' if "LARCENY" in row[2]
        else
        ).add_to(example_map1)

example_map1
```
